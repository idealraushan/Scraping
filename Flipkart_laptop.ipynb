```python
# import libraries

import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# paste the url
# URL of flipkart laptop price under 60000

url = "https://www.flipkart.com/search?q=laptop+under+60000&as=on&as-show=on&otracker=AS_Query_OrganicAutoSuggest_8_12_na_na_na&otracker1=AS_Query_OrganicAutoSuggest_8_12_na_na_na&as-pos=8&as-type=RECENT&suggestionId=laptop+under+60000&requestId=f3e5274c-e196-4a3c-9705-6f3233fdf4a5&as-searchtext=laptop%20under"
r = requests.get(url)
print(r.status_code)
```

```python
soup = BeautifulSoup(r.text,"lxml")
print(soup)
```

```python
model = (soup.find("div",{"class":"_4rR01T"}))
print(model.string)
```

```python
# Company or Brand name of Laptops

Brand_names = soup.find_all("div", class_ = "_4rR01T")
#print(names)
product = []
for i in Brand_names:
    name = i.text
    product.append(name)
print(product)
    
```

```python
# Over view and description of Laptops

desc = soup.find_all("ul" , class_ = "_1xgFaf")
product_desc = []

for i in desc:
    description = i.text
    product_desc.append(description)
print(product_desc)
```

```python
# discounted price of the given laptop

prices = soup.find_all("div" , class_ = "_30jeq3 _1_WHN1")
product_price = []
for i in prices:
    price = i.text
    product_price.append(price)
print(product_price)
                      
```

```python
# Original price of Laptops

org_prices = soup.find_all("div" , class_ = "_3I9_wc _27UcVY")
product_org_price = []
for i in org_prices:
    org_price = i.text
    product_org_price.append(org_price)
print(product_org_price)
```

```python
# Rating of the laptop

ratings = soup.find_all("div" , class_ = "_3LWZlK")
product_rating = []
for i in ratings:
    rating = i.text
    product_rating.append(rating)
print(product_rating)
```

```python
# Creating a dataframe of product

df = pd.DataFrame({"Brand name":product,"Description":product_desc,"Original price":product_org_price ,"Discounted price":product_price})
print(df)
```

```python
# Export the Dataframe to csv file

df.to_csv("Flipkart_laptop_details.csv")
```
